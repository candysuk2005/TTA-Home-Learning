```python
import numpy as np
import pandas as pd

#Read CSV file of hoilday destinations
data = pd.read_csv("top_dest_hotel.csv")

data
```

```python
# 1. How many rows and columns are there in your file?
data.shape
```

```python
# 2. Print row 3-8 ( using iloc/loc).
data.iloc[3:9]
```

```python
# 3. Find the mean number of all-inclusive hotels across all destinations.
meanHotel = data['no_inclv_hotel'].mean()
meanHotel
```

```python
# 4. Find the lowest scoring destination.
min = data['score'].min()
minData = data[ data['score']== min]
minData
```

```python
# 5. Find the highest scoring destination.
max = data['score'].max()
maxData = data[ data['score']== max]
maxData
```

```python
# 6. Find all the destinations where there are more than 9 all-inclusive hotels.
filter = data['no_inclv_hotel'] > meanHotel
data[filter]
```

```python
# 7. Filter the data by score above 8.
filter = data['score'] > 8
data[filter]

```

```python
# 8. Filter the data score below 2 ( I need to know if these destinations should be removed or there is a problem)
filter = data['score'] > 2
data[filter]
```

```python
# Extension
# 1. Is there a correlation between number of all-inclusive hotels and score?
data['no_inclv_hotel'].corr(data['score'])
```

```python
# 2. Create a data visualisation diagram to show destination and highest scores?
import matplotlib.pyplot as plt

filter = data['score'] > 2
myData = data[filter]
sorted_data = myData.sort_values('score')
x = sorted_data['destination']
y = sorted_data['score']

plt.figure(figsize= (25, 10), linewidth= 5)
plt.title('Tripadvisor recommanded destinations')
plt.xlabel('Destination')
plt.ylabel('score')
plt.grid()
plt.scatter(x, y)
```
